
## Step 1: Import Libraries
We use Pandas for data handling, Matplotlib/Seaborn for visualization.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Clean Dataset
We use the preprocessed dataset from Week‑2.


```python
df = pd.read_csv("../Week2_DataAnalysis/clean_dataset.csv")
```

## Step 3: Correlation Analysis
Check relationships between numeric features using a heatmap.


```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64','float64'])

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

## Step 4: Distribution Analysis
Visualize distributions of Age, Salary, and PerformanceScore.


```python
sns.histplot(df['Age'], bins=10, kde=True)
plt.show()

sns.histplot(df['Salary'], bins=20, kde=True)
plt.show()

sns.boxplot(x=df['PerformanceScore'])
plt.show()
```

## Step 5: Department vs Performance
Compare performance scores across departments.


```python
sns.boxplot(x='Department_IT', y='PerformanceScore', data=df)
plt.show()
```

## Step 6: Scatterplots
Check Salary vs Age and Salary vs Performance.


```python
sns.scatterplot(x='Age', y='Salary', data=df)
plt.show()

sns.scatterplot(x='PerformanceScore', y='Salary', data=df)
plt.show()
```

## Step 7: Insights
Summarize 5 key findings from the analysis:

- Younger employees tend to have lower salaries.

- PerformanceScore shows mild positive correlation with Salary.

- IT department employees have higher average performance scores.

- Longer tenure (TenureDays) does not strongly affect Salary.

- Salary distribution is skewed, with a few very high earners.
