## Step 1: Import Libraries
We use Pandas for data handling, Matplotlib/Seaborn for visualization.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Dataset
Load the raw dataset into a DataFrame.


```python
df = pd.read_csv("sample.csv")
```

## Step 3: Data Summary
Check first rows, statistics, and column info.


```python
df.head()
df.describe()
df.info()
```

## Step 4: Missing Value Detection
Identify columns with missing values.


```python
df.isnull().sum()
```

## Step 5: Data Type Analysis
Check column data types.


```python
df.dtypes
```

## Step 6: Visualization
Histogram for Age distribution and correlation heatmap for numeric features.


```python
sns.histplot(df['Age'], bins=20, kde=True)
plt.show()

numeric_df = df.select_dtypes(include=['int64','float64'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```
